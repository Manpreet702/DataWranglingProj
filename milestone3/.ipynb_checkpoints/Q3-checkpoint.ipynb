```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 12

base = "hf://datasets/hao-li/AIDev/"

# -------------------------------------------------------
# 1. Load Relevant Tables
# -------------------------------------------------------
pulls = pd.read_parquet(base + "all_pull_request.parquet")
repos = pd.read_parquet(base + "all_repository.parquet")

print("Loaded all_pull_request and all_repository.")


# -------------------------------------------------------
# 2. Filter to only Agentic PRs
# agent column indicates which coding agent submitted the PR
# -------------------------------------------------------
agentic_prs = pulls[pulls["agent"].notna()].copy()

print(f"Total Agentic PRs: {len(agentic_prs)}")


# -------------------------------------------------------
# 3. Join PRs with Repository Language
# pull_request.repo_id  <-> repository.id
# -------------------------------------------------------
merged_df = agentic_prs.merge(
    repos[["id", "language"]],
    left_on="repo_id",
    right_on="id",
    how="left"
)

merged_df.rename(columns={"id_x": "pr_id", "id_y": "repo_id_actual"}, inplace=True)


# -------------------------------------------------------
# 4. Define "Closed Without Merge"
# state == 'closed' AND merged_at is NULL
# -------------------------------------------------------
merged_df["closed_without_merge"] = (
    (merged_df["state"] == "closed") &
    (merged_df["merged_at"].isna())
)


# -------------------------------------------------------
# 5. Compute rejection rate per language
# -------------------------------------------------------
lang_stats = (
    merged_df.groupby("language")
    .agg(
        total_prs=("pr_id", "count"),
        rejected=("closed_without_merge", "sum")
    )
    .reset_index()
)

lang_stats["rejection_rate"] = lang_stats["rejected"] / lang_stats["total_prs"] * 100


print("Language rejection statistics:")
display(lang_stats.sort_values("rejection_rate", ascending=False))


# -------------------------------------------------------
# 6. Chi-square test:
# Does rejection rate differ by language?
# -------------------------------------------------------
contingency = (
    merged_df.groupby("language")["closed_without_merge"]
    .value_counts()
    .unstack(fill_value=0)
)

chi2, p, dof, expected = chi2_contingency(contingency)

print("\nChi-Square Test for Independence:")
print(f"Chi2 statistic: {chi2:.4f}")
print(f"p-value:        {p:.4e}")
print(f"Degrees of freedom: {dof}")
print("\nInterpretation:")
print("- p < 0.05 suggests that programming language significantly influences the likelihood")
print("  that agent-submitted PRs are closed without merging.")


# -------------------------------------------------------
# 7. Bar chart of rejection rates by language
# -------------------------------------------------------
top10_langs = lang_stats.sort_values("rejection_rate", ascending=False).head(10)

plt.figure(figsize=(12, 7))
sns.barplot(
    data=top10_langs,
    x="language",
    y="rejection_rate",
    palette="viridis"
)

plt.title("Top 10 Languages by Rejection Rate (Closed Without Merge)\nAgentic PRs Only")
plt.ylabel("Rejection Rate (%)")
plt.xlabel("Primary Language")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
